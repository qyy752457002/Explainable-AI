```python
import matplotlib.pyplot as plt
import numpy as np

# Data
Models = ['Model 29', 'Model 30', 'Model 31', 'Model 32', 'Model 33', 'Model 34', 
          'Model 35', 'Model 36', 'Model 37', 'Model 38', 'Model 39', 'Model 40']

trace_fitness = [1.00, 1.00, 0.09, 1.00, 1.00, 1.00, 0.15, 1.00, 1.00, 1.00, 0.21, 1.00]
move_model_fitness = [1.00, 1.00, 0.91, 1.00, 1.00, 1.00, 0.98, 1.00, 1.00, 1.00, 0.99, 1.00]
move_log_fitness = [1.00, 1.00, 0.09, 1.00, 1.00, 1.00, 0.13, 1.00, 1.00, 1.00, 0.19, 1.00]

# Create the plot
plt.figure(figsize=(12, 6))

plt.plot(Models, trace_fitness, marker='o', label='Trace Fitness', linewidth=2, markersize=8)
plt.plot(Models, move_model_fitness, marker='s', label='Move-Model Fitness', linewidth=2, markersize=8)
plt.plot(Models, move_log_fitness, marker='^', label='Move-Log Fitness', linewidth=2, markersize=8)

plt.xlabel('Model')
plt.ylabel('Fitness Value')
plt.title('Fitness Metrics Across Model 29 through Model 40')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

# Show the plot
plt.show()
```

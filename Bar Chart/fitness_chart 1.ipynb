```python
import matplotlib.pyplot as plt

# Data
Models = ['Model 5', 'Model 6', 'Model 7', 'Model 8', 'Model 9', 'Model 10', 
          'Model 11', 'Model 12', 'Model 13', 'Model 14', 'Model 15', 'Model 16']

trace_fitness = [0.10, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 0.15, 1.00, 0.12, 1.00]
move_model_fitness = [0.77, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 0.91, 1.00, 0.94, 1.00]
move_log_fitness = [0.10, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 1.00, 0.15, 1.00, 0.11, 1.00]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(Models, trace_fitness, marker='o', label='Trace Fitness', linewidth=2, markersize=8)
plt.plot(Models, move_model_fitness, marker='s', label='Move-Model Fitness', linewidth=2, markersize=8)
plt.plot(Models, move_log_fitness, marker='^', label='Move-Log Fitness', linewidth=2, markersize=8)

plt.title('Fitness Metrics Across Model 5 through Model 16')
plt.xlabel('Model')
plt.ylabel('Fitness Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

# Show the plot
plt.show()
```
